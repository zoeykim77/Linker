```python
import json

def get_course_id(event_properties):
    try:
        properties = json.loads(event_properties)
        return properties.get('course_id')
    except json.JSONDecodeError:
        return None

lesson['course_id'] = lesson['event_properties'].apply(get_course_id)

l101_lessons = lesson[lesson['course_id'] == 'L101']
display(l101_lessons)
```

```python
import json

def get_lesson_id(event_properties):
    try:
        properties = json.loads(event_properties)
        return properties.get('lesson_id')
    except json.JSONDecodeError:
        return None

l101_lessons['lesson_id'] = l101_lessons['event_properties'].apply(get_lesson_id)
unique_lesson_ids = l101_lessons['lesson_id'].unique()
display(unique_lesson_ids)
```
