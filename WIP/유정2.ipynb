```python
counts = chat_events.groupby('channel')['status'].value_counts()
sort = ['open', 'in_progress', 'resolved', 'closed']
counts = counts.reindex(index=sort, level='status')
counts
```

```python
abc= chat_events['channel'].value_counts()
abc
```

```python
counts_df = counts.reset_index(name='count')
display(counts_df)
```

```python
# 채널별 파이 그래프 생성
channels = counts_df['channel'].unique()

# 사용자 정의 색상 팔레트 (resolved, closed: 노란색 계열, open, in_progress: 초록색 계열)
status_colors = {
    'open': '#00bf63',
    'in_progress': '#aaf386',
    'resolved': '#00bf63',
    'closed': '#aaf386'
}


fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.flatten() # 2x2 배열을 1차원으로 변환

for i, channel in enumerate(channels):
    channel_data = counts_df[counts_df['channel'] == channel]
    status_counts = channel_data.set_index('status')['count']

    # 상태 순서 및 색상 매핑 유지
    ordered_statuses = ['open', 'in_progress', 'resolved', 'closed']
    status_counts = status_counts.reindex(ordered_statuses).fillna(0)
    colors = [status_colors[status] for status in status_counts.index]


    axes[i].pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
    axes[i].set_title(f'{channel} 채널 상담 상태 분포')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 한글 폰트 설정
# 시스템에 설치된 나눔고딕 폰트 경로를 찾아서 사용
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font_name
plt.rcParams['axes.unicode_minus'] = False

# 월별로 event_name 분포 파이 차트 시각화
months = monthly_event_counts['event_timestamp'].unique()

for month in months:
    monthly_data = monthly_event_counts[monthly_event_counts['event_timestamp'] == month]
    event_counts = monthly_data.set_index('event_name')['count'].sort_values(ascending=False) # 값 기준으로 내림차순 정렬

    # 값 순서에 따라 색상 팔레트 적용 (값이 클수록 진한 색)
    colors = sns.color_palette('Greens', len(event_counts))


    plt.figure(figsize=(8, 8))
    plt.pie(event_counts, labels=event_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
    plt.title(f'{month} 월별 이벤트 분포')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```
